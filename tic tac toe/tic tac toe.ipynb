```python
def check_if_won(board , p , x):
#     diagonal check, left to right
    if x % 4 == 0:
        if board[0] == board[4] == board[8] == p:
            return True
#     horizontal check    
    i = x // 3 * 3
    if board[i] == board[i+1] == board[i+2] == p:
        return True
#     vertical check
    #     to get to the first row in that column
    #     i has the row of x from previous check
    if i == 3:          # 2nd row
        i = x - 3
    elif i == 6:        # 3rd row
        i = x - 6
    else:               # 1st row
        i = x
    if board[i] == board[i+3] == board[i+6] == p:
        return True;
#     diagonal check, right to left
    if x == 2 or x == 4 or x == 6:
        if board[2] == board[4] == board[6] == p:
            return True
    return False
```

```python
def board_filled(board):
    for i in range(9):
        if str(i) in board:
            return False
    return True
```

```python
def display_board(board):
    print("\t ===========")
    line = "\t"
    for i in range(9):
        line = line + '| ' + board[i] + ' '
        if (i+1) % 3 != 0:
            continue;
        line= line + '|'
        print(line)
        line = '\t'
        if i//3 < 2:
            print("\t|-----------|")
    print("\t ===========")
```

```python
w1 = 0
w2 = 0
def game_play(name1 , name2):
    board = ['1' , '2' , '3' , '4' , '5' , '6' , '7' , '8' , '9']
    print(name1 , " : Select your Symbol from the following:")
    print("\tX or +")
    while True:
        p1 = input()
        if p1 == 'X':
            p2 = '+'
            break
        elif p1 == '+':
            p2 = 'X'
            break
        else:
            print("\tPlease select from the options given!")
    print(name1 + " : " + p1 + "\n" + name2 + " : " + p2)
    i = 1
    while True:
        print("Chance: " , i)
        display_board(board)
        print(name1 + " : Where would you like to mark your symbol?")
        while True:
            x = input()
            if x in board:
                x = int(x)
                board[x-1] = p1;
                break;
            else:
                print("Please enter the integers shown on the board")
        display_board(board)
        if check_if_won(board , p1 , x-1):
            print(name1 + " won!!")
            global w1
            w1 += 1
            break
        if board_filled(board):
            print("Oooops! The board is filled!")
            break
        print(name2 + " : Where would you like to mark your symbol?")
        while True:
            x = input()
            if x in board:
                x = int(x)
                board[x-1] = p2;
                break;
            else:
                print("Please enter the integers shown on the board")
        display_board(board)
        if check_if_won(board , p2 , x-1):
            print(name2 + " won!!")
            global w2
            w2 += 1
            break
        i += 1
```

```python
import datetime

def show_stats(name1 , name2):
    global w1
    global w2
    print("\t" + name1 + " have won " , w1 , " rounds")
    print("\t" + name2 + " have won " , w2 , " rounds")
    if w1 > w2:
        print("Congratulations " + name1 + "!!! You won this game!")
    elif w2 > w1:
        print("Congratulations " + name2 + "!!! You won this game!")
    else:
        print("Oh! It was a tie then!!")
    f = open("tic_tac_toe.txt" , "at")
    line = name1 + " : " + str(w1) + " v/s " + name2 + " : " + str(w2) + "\t" + str(datetime.datetime.now()) + "\n"
    f.write(line)
    f.close()
    w1 = 0
    w2 = 0
```

```python
def tic_tac_toe():
    while True:
        print("Enter your game names")
        name1 = input("\tPlayer 1: ")
        name2 = input("\tPlayer 2: ")
        while True:
            game_play(name1 , name2)
            print("Do you want to play another round? Enter 'y' to confirm...")
            x = input()
            if x != 'y' and x != 'Y':
                break
        show_stats(name1 , name2)
        print("Enter 'y' to check all the scores so far...")
        x = input()
        if x == 'y' or x == 'Y':
            f = open("tic_tac_toe.txt" , "r")
            for x in f:
                print(x)
            f.close()
        print("Do you want to play a new game? Enter 'y' to confirm...")
        x = input()
        if x != 'y' and x != 'Y':
            break
```
