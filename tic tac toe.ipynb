```python
def check_if_won(board , p , x):
#     diagonal check, left to right
    if x % 4 == 0:
        if board[0] == board[4] == board[8] == p:
            return True
#     horizontal check    
    i = x // 3 * 3
    if board[i] == board[i+1] == board[i+2] == p:
        return True
#     vertical check
    #     to get to the first row in that column
    #     i has the row of x from previous check
    if i == 3:          # 2nd row
        i = x - 3
    elif i == 6:        # 3rd row
        i = x - 6
    else:               # 1st row
        i = x
    if board[i] == board[i+3] == board[i+6] == p:
        return True;
#     diagonal check, right to left
    if x == 2 or x == 4 or x == 6:
        if board[2] == board[4] == board[6] == p:
            return True
    return False
```

```python
def board_filled(board):
    for i in range(9):
        if str(i) in board:
            return False
    return True
```

```python
def display_board(board):
    print("\t ===========")
    line = "\t"
    for i in range(9):
        line = line + '| ' + board[i] + ' '
        if (i+1) % 3 != 0:
            continue;
        line= line + '|'
        print(line)
        line = '\t'
        if i//3 < 2:
            print("\t|-----------|")
    print("\t ===========")
```

```python
game_play = True
while game_play == True:
    board = ['1' , '2' , '3' , '4' , '5' , '6' , '7' , '8' , '9']
    print("Player 1: Select your Symbol from the following:")
    print("\tX or +")
    while True:
        p1 = input()
        if p1 == 'X':
            p2 = '+'
            break
        elif p1 == '+':
            p2 = 'X'
            break
        else:
            print("\tPlease select from the options given!")
    print("Player 1 : " + p1 + "\nPlayer 2 : " + p2)
    i = 1
    while True:
        print("Attempt " , i)
        display_board(board)
        print("Player 1: Where would you like to mark your symbol?")
        while True:
            x = input()
            if x in board:
                x = int(x)
                board[x-1] = p1;
                break;
            else:
                print("Please enter the integers shown on the board")
        display_board(board)
        if check_if_won(board , p1 , x-1):
            print("Player 1 won!!")
            break
        if board_filled(board):
            print("Oooops! The board is filled!")
            break
        print("Player 2: Where would you like to mark your symbol?")
        while True:
            x = input()
            if x in board:
                x = int(x)
                board[x-1] = p2;
                break;
            else:
                print("Please enter the integers shown on the board")
        display_board(board)
        if check_if_won(board , p2 , x-1):
            print("Player 2 won!!")
            break
        i += 1
    game_play = False
```
